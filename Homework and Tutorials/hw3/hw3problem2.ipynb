```python
# Import all modules

import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson     # Poisson function
import seaborn as sb                # Seaborn library to generate frequency plots
```

Generate Poisson distributions (different sets of y-values from the same x-values)

```python
# Import Poisson distribution

# Choose your sample size n
n = 100

# Generate Poisson distributions
y1 = poisson.rvs(mu=1, size=n)
y2 = poisson.rvs(mu=2, size=n)
y4 = poisson.rvs(mu=4, size=n)
y10 = poisson.rvs(mu=10, size=n)
print(np.mean(y1))
print(np.mean(y2))
print(np.mean(y4))
print(np.mean(y10))
```

Generate frequency plots

```python
# Generate frequency plots

# Initialize the number of subplots (rows, columns)
# figure, axis = plt.subplots(4)

sb.distplot(y1,
                    kde=True,
                    hist_kws={"linewidth": 25, 'alpha': 1},
                    label="$\mu=1$"
)
sb.distplot(y2,
                    kde=True,
                    hist_kws={"linewidth": 25, 'alpha': 1},
                    label="$\mu=2$"
)
sb.distplot(y4,
                    kde=True,
                    hist_kws={"linewidth": 25, 'alpha': 1},
                    label="$\mu=4$"
)
sb.distplot(y10,
                    kde=True,
                    hist_kws={"linewidth": 25, 'alpha': 1},
                    label="$\mu=10$"
)
# Display the plot
plt.legend(loc='best')
plt.xlabel('Poisson')
plt.ylabel('Frequency')

# Save the plot
plt.savefig('./hw3problem2c.png')

plt.show()
```
