```python
# Import all modules

import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson     # Poisson function
from scipy.stats import norm        # Gaussian function
import seaborn as sb                # Seaborn library to generate frequency plots
```

Generate Poisson distributions (different sets of y-values from the same x-values)

```python
# Import Poisson distribution

# Choose your x-range
xmax = 30
x = np.arange(xmax)

# Generate Poisson distributions
y1 = poisson.pmf(x,mu=1)
y2 = poisson.pmf(x,mu=2)
y4 = poisson.pmf(x,mu=4)
y10 = poisson.pmf(x,mu=10)

# mu=1
color1='r'
plt.plot(x,y1,'r-', label='$\mu=1$')
# plt.vlines(x,0,y1,colors='r',linestyles='-')
plt.bar(x,y1,color=color1)

# mu=2
color2='y'
plt.plot(x,y2,'y-', label='$\mu=2$')
# plt.vlines(x,0,y2,colors=color2,linestyles='-')
plt.bar(x,y2,color=color2)

# mu=4
color4='g'
plt.plot(x,y4,'g-', label='$\mu=4$')
# plt.vlines(x,0,y4,colors='g',linestyles='-')
plt.bar(x,y4,color=color4)

# mu=10
color10='b'
plt.plot(x,y10,'b-', label='$\mu=10$')
# plt.vlines(x,0,y10,colors=color10,linestyles='-')
plt.bar(x,y10,color=color10)

# rv = poisson(mu)
# ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
#         label='frozen pmf')

# Display the plot
plt.legend(loc='best')
plt.title('Poisson Distribution for $\mu=1,2,4,10$')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Save the plot
plt.savefig('./hw3problem2c_alt.png')

plt.show()
```

Generate Poisson distribution and Gaussian distribution for $\mu=30$

```python
# Choose your x-range
xmin = 10
xmax = 50
xstep = 1
x = np.arange(xmin,xmax,xstep)

# Choose your mean
mu = 30

# Generate Poisson distribution
poiss_30 = poisson.pmf(x,mu=mu)

# Generate Gaussian distribution
sd = np.sqrt(30)
gauss_30 = norm.pdf(x,loc=mu, scale=sd)

# Plot Poisson distribution
color1='r'
plt.plot(x,poiss_30,'ro-',label='Poisson')
# plt.bar(x,poiss_30,color=color1)

# Plot Gaussian distribution
color2='y'
plt.plot(x,gauss_30,'b^-',label='Gaussian')

# Add labels
plt.legend(loc='best')
plt.title('Poisson vs Gaussian Distribution for $\mu=30$')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Save figure
plt.savefig('./hw3problem2d.png')

plt.show()
```
